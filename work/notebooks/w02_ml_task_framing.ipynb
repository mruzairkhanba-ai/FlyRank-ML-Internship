# 1. My Lane as an ML Task

**Task Type:** Classification

The goal of this project is to classify whether a content page needs to be refreshed or not.

Each row in the dataset represents one content page with information such as search volume, clicks, impressions, average position, engagement, and trend direction.

The model will help identify pages that should be prioritized for updates.

# Target or Proxy

The dataset does not contain a direct target column called `needs_refresh`. Therefore, I created a proxy target.

A page is considered to need a refresh if:
- It has fewer than 100 clicks in the last 90 days.
- Its average search position is greater than 20.

These conditions indicate that the content has low engagement and poor search visibility.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("/content/content_refresh_anonymized.csv")

print("Available columns in df:", df.columns)

# Example proxy target
df["needs_refresh"] = (
    (df["organic_clicks"] < 100) &
    (df["position"] > 20)
).astype(int)

df[["organic_clicks","position","needs_refresh"]].head()
```

# Success Metric

The model will be evaluated using:

- Accuracy
- Precision
- Recall
- F1 Score

These metrics measure how well the model identifies pages that need refreshing.

# The Unit of Analysis

One row in the dataset represents one content page.

Each page contains information such as:

The model will use these features to predict whether a page needs refreshing.

# Why ML Beats a Fixed Rule Here

A fixed rule can only check one or two conditions, such as low clicks or poor search ranking.

Machine learning is better because it can learn patterns from many features at the same time, including impressions, CTR, search volume, content age, engagement, and average position.

This makes predictions more accurate and adaptable than using a single fixed rule.

# Real Content Action

The model supports a real business decision.

If the model predicts that a page needs refreshing, the content team can review and update that page.

If the model predicts that a page does not need refreshing, the page can continue performing without immediate changes.

This helps prioritize work and use resources more efficiently.
