## 1. My lane (or freestyle) and why





I chose the Content Refresh lane because this dataset contains content performance metrics such as impressions, clicks, pageviews, sessions, and users. My goal is to identify which pages should be refreshed first. This helps content teams focus their effort on pages that are more likely to benefit from an update.

## 2. The question: decision, action, cost of a wrong call



Research Question:
Which content pages should be prioritized for refresh?

Decision:
Decide which pages should be refreshed first.

Unit of analysis:
One content page.

Output:
A recommendation that identifies high-priority pages for refresh.

Action:
The content team refreshes the recommended pages.

Cost of a wrong recommendation:
Refreshing the wrong pages wastes time and resources while important pages continue to perform poorly.

Why ML can help:
Machine learning can identify patterns in historical content performance and support better prioritization decisions.
